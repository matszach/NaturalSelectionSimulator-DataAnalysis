```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from statistics import mean
```

```python
def plot_size_vs_speed_in_time(session):
    
    # set-up the plot
    plt.title(f'Creature traits in time, session {session}')
    plt.xlim(0,520)
    plt.xlabel('size')
    plt.ylim(0,520)
    plt.ylabel('speed')
    
    # amount of samples in directory
    amt = len(os.listdir(f'datasets/data/session_{session}/creature_data'))
    
    # for each sample
    for i in range(1,amt):
        df = pd.read_csv(f'datasets/data/session_{session}/creature_data/creature_data_{i}.csv')
        x = df.iloc[:,1]
        y = df.iloc[:,2]
        plt.scatter(x, y, c=[(0,(amt-i)/(amt*1.2),1) for j in range(len(x))])
        
    # show plot
    plt.show()
    
```

```python
for i in range(1,11):
    plot_size_vs_speed_in_time(i)
```

```python
def plot_population_in_time(session):
    
    # set-up the plot
    plt.title(f'Creature and food population in time, session {session}')
    plt.xlabel('sample')
    plt.ylabel('number')
    
    # read data
    df = pd.read_csv(f'datasets/data/session_{session}/session_data.csv')
    creatures = df.iloc[:,1]
    food = df.iloc[:,2]
    
    # plot populations
    plt.plot(creatures, 'purple', linewidth=2)
    plt.plot(food, 'green', linewidth=2)
    
    # show plot
    plt.show()
    
```

```python
for i in range(1,11):
    plot_population_in_time(i)
```

```python
def plot_avg_traits_in_time(session):
    
    # plottable lists
    speeds = []
    sizes = []
    
    # amount of samples in directory
    amt = len(os.listdir(f'datasets/data/session_{session}/creature_data'))
    
    # for each sample
    for i in range(1,amt):
        df = pd.read_csv(f'datasets/data/session_{session}/creature_data/creature_data_{i}.csv')
        sizes.append(mean(df.iloc[:,1]))
        speeds.append(mean(df.iloc[:,2]))
    
    plt.plot(speeds, 'red')
    plt.plot(sizes, 'green')
    plt.show()
```

```python
for i in range(1,11):
    plot_avg_traits_in_time(i)
```
